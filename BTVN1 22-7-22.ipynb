```python
import turtle
import math

chieudai = float(input("Nhập chiều dài lá cờ: "))
a = chieudai/3
chieurong = 2*a
print("Chiều rộng lá cờ là: ", chieurong)
dinhngoisao = chieurong/6

Xdinhngoisao = 0
Ydinhngoisao = (chieurong/2) - dinhngoisao
canhvuongdai = (4*a)/3
tanC = math.tan((math.pi)/10) #18 độ = (math.pi)/10
canhvuongngan = ((4*a)/3)*tanC
canhhuyen = math.sqrt(canhvuongdai**2 + canhvuongngan**2)

print(a)
print(canhvuongdai)
print(canhvuongngan)
print(canhhuyen)

t = turtle.Turtle()
t.shape()
t.hideturtle()
t.pensize(1)
t.speed(10)

t.penup()
t.goto(-chieudai/2, chieurong/2)
t.pendown()

t.color('#FECB00')
t.fillcolor('#FECB00')
t.begin_fill()
t.forward(chieudai)
t.right(90)
t.forward(chieurong/3)
t.right(90)
t.forward(chieudai)
t.right(90)
t.forward(chieurong/3)
t.backward(chieurong/3)
t.end_fill()

t.color('#34B233')
t.fillcolor('#34B233')
t.begin_fill()
t.right(90)
t.forward(chieudai)
t.right(90)
t.forward(chieurong/3)
t.right(90)
t.forward(chieudai)
t.right(90)
t.forward(chieurong/3)
t.backward(chieurong/3)
t.end_fill()

t.color('#EA2839')
t.fillcolor('#EA2839')
t.begin_fill()
t.right(90)
t.forward(chieudai)
t.right(90)
t.forward(chieurong/3)
t.right(90)
t.forward(chieudai)
t.right(90)
t.forward(chieurong/3)
t.end_fill()

t.penup()
t.goto(Xdinhngoisao, Ydinhngoisao)
t.pendown()
t.left(162) # 180 - (180-144)/2
t.color('#FFFFFF')
t.fillcolor('#FFFFFF')
t.begin_fill()
t.forward(canhhuyen)
t.left(144)
t.forward(canhhuyen)
t.left(144)
t.forward(canhhuyen)
t.left(144)
t.forward(canhhuyen)
t.end_fill()

turtle.done()
```
