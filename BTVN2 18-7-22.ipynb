```python
import turtle

print("Nhập tọa độ điểm A: ")
xA = int(input("xA = "))
yA = int(input("yA = "))

print("Nhập tọa độ điểm B: ")
xB = int(input("xB = "))
yB = int(input("yB = "))

print("Nhập tọa độ điểm C: ")
xC = int(input("xC = "))
yC = int(input("yC = "))

xD1 = xA - xB + xC
yD1 = yA - yB + yC
print("Tọa độ điểm D1","(", xD1,",", yD1,")")

xD2 = xA - xC + xB
yD2 = yA - yC + yB
print("Tọa độ điểm D2","(", xD2,",", yD2,")")

xD3 = xB - xA + xC
yD3 = yB - yA + yC
print("Tọa độ điểm D3","(", xD3,",", yD3,")")

t = turtle.Turtle()
t.shape()
t.hideturtle()
t.pensize(3)

t.penup()
t.goto(xA*50, yA*50)
t.dot(15,"green")
t.pendown()
t.goto(xB*50, yB*50)
t.dot(15,"green")
t.goto(xC*50, yC*50)
t.dot(15,"green")
t.goto(xD1*50, yD1*50)
t.dot(15,"blue")
t.goto(xA*50, yA*50)

t.penup()
t.goto(xA*50, yA*50)
t.dot(15,"green")
t.pendown()
t.goto(xC*50, yC*50)
t.dot(15,"green")
t.goto(xB*50, yB*50)
t.dot(15,"green")
t.goto(xD2*50, yD2*50)
t.dot(15,"blue")
t.goto(xA*50, yA*50)

t.penup()
t.goto(xB*50, yB*50)
t.dot(15,"green")
t.pendown()
t.goto(xA*50, yA*50)
t.dot(15,"green")
t.goto(xC*50, yC*50)
t.dot(15,"green")
t.goto(xD3*50, yD3*50)
t.dot(15,"blue")
t.goto(xB*50, yB*50)

turtle.done()
```
