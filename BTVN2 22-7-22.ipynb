```python
import turtle
import math

print("Nhập tọa độ điểm A:")
xA = int(input("xA = "))
yA = int(input("yA = "))
print("Nhập tọa độ điểm B:")
xB = int(input("xB = "))
yB = int(input("yB = "))
print("Nhập tọa độ điểm C:")
xC = int(input("xC = "))
yC = int(input("yC = "))

xAB = xB - xA
yAB = yB - yA
xAC = xC - xA
yAC = yC - yA
xBC = xB - xC
yBC = yB - yC

tamgiac = (xAB != yAB) or (xAC != yAC)
print("Tam giác = ", tamgiac)

AB = math.sqrt((xB-xA)**2+(yB-yA)**2)
BC = math.sqrt((xC-xB)**2+(yC-yB)**2)
AC = math.sqrt((xC-xA)**2+(yC-yA)**2)

cosA = (AB**2+AC**2-BC**2)/(2*AB*AC)
cosB = (AB**2+BC**2-AC**2)/(2*AB*BC)
cosC = (BC**2+AC**2-AB**2)/(2*BC*AC)

gocA = math.degrees(math.acos(cosA))
gocB = math.degrees(math.acos(cosB))
gocC = math.degrees(math.acos(cosC))

# Độ dài các đường cao
caoA = AB * math.sin(math.radians(gocB))
caoB = AB * math.sin(math.radians(gocA))
caoC = BC * math.sin(math.radians(gocC))

print("Độ dài đường cao từ góc A:", caoA)
print("Độ dài đường cao từ góc B:", caoB)
print("Độ dài đường cao từ góc C:", caoC)

# Độ dài các đường trung tuyến
nuaBC = BC/2
ttA = math.sqrt(AC**2 + nuaBC**2 - 2*AC*nuaBC*cosC)
nuaAC = AC/2
ttB = math.sqrt(AB**2 + nuaAC**2 - 2*AB*nuaAC*cosA)
nuaAB =AB/2
ttC = math.sqrt(BC**2 + nuaAB**2 - 2*BC*nuaAB*cosB)

print("Độ dài đường trung tuyến từ góc A:", ttA)
print("Độ dài đường trung tuyến từ góc B:", ttB)
print("Độ dài đường trung tuyến từ góc C:", ttC)

# Tọa độ các trọng tâm
xTRONGTAM = (xA + xB + xC)/3
yTRONGTAM = (yA + yB + yC)/3
print("Tọa độ trọng tâm tam giác là:", xTRONGTAM, yTRONGTAM)

# Tọa độ trực tâm
D = (xAB*yAC) - (xAC*yAB)
Dx = yAC*(xAB*xC + yAB*yC) - yAB*(xAC*xB + yAC*yB)
Dy = xAB*(xAC*xB + yAC*yB) - xAC*(xAB*xC + yAB*yC)

xTRUCTAM = Dx/D
yTRUCTAM = Dy/D

print("Tọa độ trực tâm tam giác là:", xTRUCTAM, yTRUCTAM)

t = turtle.Turtle()
t.shape()
t.hideturtle()
t.pensize(3)
t.color("black")

t.penup()
t.goto(xA, yA)
t.pendown()
t.dot(15,"Blue")
t.goto(xB, yB)
t.dot(15,"Blue")
t.goto(xC, yC)
t.dot(15,"Blue")
t.goto(xA, yA)

# Vẽ trọng tâm
t.penup()
t.goto(xTRONGTAM, yTRONGTAM)
t.pendown()
t.dot(15,"Red")
t.goto(xA, yA)
t.penup()
t.goto(xTRONGTAM, yTRONGTAM)
t.pendown()
t.goto(xB, yB)
t.penup()
t.goto(xTRONGTAM, yTRONGTAM)
t.pendown()
t.goto(xC, yC)

# Vẽ trực tâm
t.penup()
t.goto(xTRUCTAM, yTRUCTAM)
t.pendown()
t.dot(15,"Green")
t.goto(xA, yA)
t.penup()
t.goto(xTRUCTAM, yTRUCTAM)
t.pendown()
t.goto(xB, yB)
t.penup()
t.goto(xTRUCTAM, yTRUCTAM)
t.pendown()
t.goto(xC, yC)

turtle.done()
```
