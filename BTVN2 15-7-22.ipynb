```python
import turtle
import random

turtle.setup(800, 600)
t = turtle.Turtle()
t.shape("turtle")
t.hideturtle()
t.color("black")
t.pensize(5)
t.speed(30)
turtle.bgcolor("lightskyblue")

# Vẽ thân
t.penup()
t.goto(-150,-250)
t.pendown()
t.fillcolor("white")
t.begin_fill()
t.circle(90,180)
t.forward(70)
t.circle(90,180)
t.forward(70)
t.end_fill()

# Vẽ đầu
t.penup()
t.goto(-160,-70)
t.pendown()
t.begin_fill()
t.circle(60,180)
t.forward(40)
t.circle(60,180)
t.forward(40)
t.end_fill()

# Vẽ mắt
t.penup()
t.goto(-210,0)
t.pendown()
t.fillcolor("black")
t.begin_fill()
t.circle(8)
t.end_fill()
t.penup()
t.goto(-140,0)
t.pendown()
t.begin_fill()
t.circle(8)
t.end_fill()

#Vẽ mũi
t.penup()
t.goto(-165,-20)
t.pendown()
t.fillcolor("#FF6600")
t.begin_fill()
t.circle(5,180)
t.forward(5)
t.circle(5,180)
t.forward(5)
t.end_fill()

#Vẽ miệng
t.penup()
t.goto(-185,-40)
t.pendown()
t.color("black")
for i in range(6):
    t.left(5)
    t.forward(5)

# Vẽ khăn
t.right(30)    
t.penup()
t.goto(-130,-80)
t.pendown()
t.fillcolor("#CC0000")
t.begin_fill()
t.circle(12,180)
t.forward(100)
t.circle(12,180)
t.forward(100)
#
t.penup()
t.goto(-210,-80)
t.pendown()
t.right(90)
t.forward(50)
t.left(90)
t.forward(30)
t.left(90)
t.forward(50)
t.end_fill()
#
t.goto(-225,-80)
t.pendown()
t.begin_fill()
t.right(180)
t.forward(70)
t.left(90)
t.forward(30)
t.left(90)
t.forward(70)
t.right(90)
t.end_fill()

#Vẽ mũ
t.penup()
t.goto(-130,35)
t.pendown()
t.fillcolor("white")
t.begin_fill()
t.circle(12,180)
t.forward(100)
t.circle(12,180)
t.forward(100)
t.end_fill()
#
t.penup()
t.goto(-233,62)
t.pendown()
t.fillcolor("#CC0000")
t.begin_fill()
t.right(90)
t.circle(90,-160)
t.left(90)
t.forward(20)
t.right(90)
t.circle(27,192)
t.end_fill()
#
t.penup()
t.goto(-78,67)
t.pendown()
t.fillcolor("white")
t.begin_fill()
t.circle(15)
t.end_fill()

#Vẽ khuy
t.penup()
t.goto(-185,-150)
t.pendown()
t.fillcolor("#FF6600")
t.begin_fill()
t.circle(8)
t.end_fill()

t.penup()
t.goto(-185,-180)
t.pendown()
t.fillcolor("#FF6600")
t.begin_fill()
t.circle(8)
t.end_fill()

t.penup()
t.goto(-185,-210)
t.pendown()
t.fillcolor("#FF6600")
t.begin_fill()
t.circle(8)
t.end_fill()

#Vẽ tay
t.penup()
t.goto(-70,-110)
t.pendown()
t.left(100)
t.forward(80)
#
t.penup()
t.goto(-37,-75)
t.pendown()
t.left(80)
t.forward(20)
#
t.penup()
t.goto(-26,-70)
t.pendown()
t.left(50)
t.backward(25)
#
t.penup()
t.goto(-298,-110)
t.pendown()
t.right(40)
t.forward(80)
#
t.penup()
t.goto(-330,-70)
t.pendown()
t.right(40)
t.forward(25)
#
t.penup()
t.goto(-340,-65)
t.pendown()
t.right(100)
t.backward(25)

#Vẽ cây thông
t.penup()
t.goto(150,70)
t.pendown()
t.fillcolor("#006600")
t.begin_fill()
t.right(52)
for i in range(3):
    t.forward(300)
    t.right(120)
t.end_fill()

t.penup()
t.goto(150,120)
t.pendown()
t.fillcolor("#006600")
t.begin_fill()
for i in range(3):
    t.forward(270)
    t.right(120)   
t.end_fill()

t.penup()
t.goto(150,170)
t.pendown()
t.fillcolor("#006600")
t.begin_fill()
for i in range(3):
    t.forward(220)
    t.right(120)   
t.end_fill()

t.penup()
t.goto(150,220)
t.pendown()
t.fillcolor("#006600")
t.begin_fill()
for i in range(3):
    t.forward(180)
    t.right(120)   
t.end_fill()

t.penup()
t.goto(120,-180)
t.pendown()
t.right(30)
t.fillcolor("#880000")
t.begin_fill()
t.forward(60)
t.left(90)
t.forward(60)
t.left(90)
t.forward(60)
t.end_fill()

t.penup()
t.goto(132,200)
t.pendown()
t.right(17)
t.color("#FFFF00")
t.fillcolor("#FFFF00")
t.begin_fill()
for i in range(5):
    t.forward(60)
    t.right(144)   
t.end_fill()

#
def circle():
    t.color("white")
    t.fillcolor("white")
    t.begin_fill()
    t.circle(3)
    t.end_fill()

for i in range(50):
    x = random.randint(-400, 400)
    y = random.randint(-300, 300)
    circle()
    t.penup()
    t.goto(x, y)
    t.pendown()
    
t = turtle.done()
```
