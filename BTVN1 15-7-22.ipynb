```python
import turtle

t = turtle.Turtle()
t.shape()
t.hideturtle()
t.pensize(1)

t.penup()
t.backward(250)
t.left(90)
t.forward(150)
t.right(90)
t.pendown()

t.color('yellow')
t.fillcolor('yellow')
t.begin_fill()
t.forward(450)
t.right(90)
t.forward(100)
t.right(90)
t.forward(450)
t.right(90)
t.forward(100)
t.backward(100)
t.end_fill()

t.color('green')
t.fillcolor('green')
t.begin_fill()
t.right(90)
t.forward(450)
t.right(90)
t.forward(100)
t.right(90)
t.forward(450)
t.right(90)
t.forward(100)
t.backward(100)
t.end_fill()

t.color('red')
t.fillcolor('red')
t.begin_fill()
t.right(90)
t.forward(450)
t.right(90)
t.forward(100)
t.right(90)
t.forward(450)
t.right(90)
t.forward(100)
t.end_fill()

t.penup()
t.forward(200)
t.pendown()

t.right(140)
t.penup()
t.forward(270)
t.left(122)
t.pendown()
t.color('white')
t.fillcolor('white')
t.begin_fill()
t.forward(150)
t.right(144)
t.forward(160)
t.right(144)
t.forward(160)
t.right(144)
t.forward(160)
t.right(144)
t.forward(160)
t.end_fill()

turtle.done()
```
