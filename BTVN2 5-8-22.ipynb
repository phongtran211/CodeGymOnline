```python
import turtle
import math

i = 0
j = 0
z = 0
k = 0
cheo = math.sqrt(120**2 + 120**2)
t = turtle.Turtle()
t.shape()
t.speed(20)
#t.hideTurtle()

def goto(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()

def khung(x, y, e):
    t.pensize(e)
    t.forward(x)
    t.left(90)
    t.forward(y)
    t.left(90)
    t.forward(x)
    t.left(90)
    t.forward(y)  
    t.left(90)

def chan():
    #goto()
    t.forward(10)
    t.right(90)
    t.forward(10)

goto(-250 + k - 10, -300 + z - 10)
khung(500, 560, 4)
goto(-250 + k, -300 + z)
khung(480, 540, 2)

while i < 8:
    z = z + 60
    goto(-250, -300 + z)
    t.forward(480)
    i = i + 1

z = 0
k = 60
goto(-250 + k, -300 + z)
t.left(90)
while j < 7:
    t.forward(240)
    goto(-250 + k, z)
    t.forward(240)
    k = k + 60
    goto(-250 + k, -300)
    j = j + 1

goto(-70, -300)
t.right(45)
t.forward(cheo)

goto(-70, 120)
t.forward(cheo)

goto(50, -300)
t.left(90)
t.forward(cheo)

goto(50, 120)
t.forward(cheo)

turtle.done()
```
